# Nelson Mandela Hybrid GraphRAG: Query

Query the hybrid graph built in Part 1 (8a) using three approaches:

1. **Semantic search** — Natural language queries via lexical-graph
2. **Structured queries** — Direct Cypher via document-graph
3. **Hybrid correlation** — Semantic results → entity matching → structured enrichment

## The Hybrid Query Pattern

```
Natural language question
    ↓ lexical-graph (vector search + graph traversal)
Semantic results (text chunks, entities)
    ↓ entity extraction from results
Entity names ("ANC", "Robben Island", "1990")
    ↓ document-graph (Cypher match on typed nodes)
Structured facts (dates, roles, relationships)
    ↓ combined answer
Rich response with both context and precision
```

**Prerequisites:**
- Part 1 (8a) completed
- `pip install graphrag-toolkit-lexical-graph graphrag-toolkit-document-graph[graphrag]`
- Neptune + OpenSearch endpoints


````python
# Setup
import json, re
from graphrag_toolkit.lexical_graph.storage import GraphStoreFactory, VectorStoreFactory
from graphrag_toolkit.lexical_graph import LexicalGraphQueryEngine
from graphrag_toolkit.document_graph.query import DocumentGraphQueryEngine

GRAPH_STORE = 'neptune-db://<your-neptune-cluster-endpoint>:8182'
VECTOR_STORE = '<your-opensearch-serverless-endpoint>'
TENANT = 'mandela'

graph_store = GraphStoreFactory.for_graph_store(GRAPH_STORE)
vector_store = VectorStoreFactory.for_vector_store(VECTOR_STORE)
gs = graph_store.__enter__()
print('Connected')
````

## 1. Semantic Search (Lexical-Graph)

Ask natural language questions. Lexical-graph finds relevant text chunks via vector
similarity, then expands context via graph traversal (entities connected to those chunks).

This works because Part 1 indexed Mandela's Wikipedia bio into the semantic layer.


````python
query_engine = LexicalGraphQueryEngine.for_traversal_based_search(graph_store, vector_store)

queries = [
    'Who was Nelson Mandela?',
    'What role did Mandela play in ending apartheid?',
    'What is the African National Congress?',
]

for q in queries:
    print(f'\nQuery: {q}')
    results = query_engine.retrieve(q)
    if isinstance(results, list) and results:
        print(f'  Results: {len(results)}')
        text = getattr(results[0].node, 'text', str(results[0]))[:200]
        print(f'  Top: {text}...')
    else:
        print(f'  {str(results)[:150]}')
````

## 2. Structured Queries (Document-Graph)

Query typed nodes directly with Cypher. Precise, fast, and schema-aware.

Document-graph nodes use tenant-scoped labels (e.g., `Eventmandela__`) so they
don't collide with lexical-graph entities in the same Neptune cluster.


````python
# Query events
event_label = f"`Event{TENANT}__`"
events = gs.execute_query(f"MATCH (n:`{event_label}`) RETURN properties(n) as props LIMIT 10")
print(f"Events in document-graph: {len(events)}")
for e in events[:5]:
    props = e.get("props", {})
    date = props.get("date", "")
    desc = props.get("description", "?")
    print(f"  [{date}] {desc}")

# Query organizations
org_label = f"`Organization{TENANT}__`"
orgs = gs.execute_query(f"MATCH (n:`{org_label}`) RETURN properties(n) as props LIMIT 10")
print(f"Organizations: {len(orgs)}")
for o in orgs[:5]:
    props = o.get("props", {})
    name = props.get("name", props.get("title", "?"))
    print(f"  {name}")
````

## 3. Hybrid Correlation

The most powerful pattern: combine semantic search results with structured data.

1. Semantic query finds relevant context from the biography
2. Extract entity names from the results (people, organizations, events)
3. Match those entities against typed document-graph nodes
4. Enrich the semantic results with structured facts (exact dates, roles, relationships)

This gives you the best of both worlds: semantic understanding + structured precision.


````python
# Semantic query + correlate entities to document-graph
q = 'What key events shaped Mandela\'s life?'
results = query_engine.retrieve(q)
print(f'Query: {q}')
print(f'Lexical results: {len(results)}\n')

# Extract entities and look them up in document-graph
for i, r in enumerate(results[:3], 1):
    meta = getattr(r.node, 'metadata', {})
    ec = meta.get('entity_contexts', {}).get('contexts', [])
    text = getattr(r.node, 'text', str(r))[:150]
    print(f'Result {i}: {text}...')

    for ctx in ec[:5]:
        for ent in ctx.get('entities', []):
            entity = ent.get('entity', {})
            value = entity.get('value', '')
            classification = entity.get('classification', '')
            if not value:
                continue
            search_term = value.split(',')[0].strip()
            found = None
            if classification in ('Event', 'Organization', 'Person'):
                for label in [f'`Event{TENANT}__`', f'`Organization{TENANT}__`']:
                    found = gs.execute_query(f"MATCH (n:`{label}`) WHERE n.name CONTAINS '{search_term}' OR n.title CONTAINS '{search_term}' RETURN properties(n) as props LIMIT 1")
                    if found:
                        break
            if found:
                props = found[0].get('props', {})
                name = props.get('name', props.get('title', value))
                print(f'    [{classification}] "{value}" -> Neptune: {name}')
    print()

print('Hybrid query complete — lexical search + document-graph correlation')
````
